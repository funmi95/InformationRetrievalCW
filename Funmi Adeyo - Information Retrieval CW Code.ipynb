```python
import requests
from bs4 import BeautifulSoup
import re

def make_soup(url):
    html = requests.get(url)
    plain = html.text
    soup = BeautifulSoup(plain, 'html.parser')  
    return soup
```

```python
#BEST
import requests
from bs4 import BeautifulSoup
import re

def crawler(seed,maxcount):
    """doc string"""
    
    Q = [seed]
    count=0
    Pub_title=[]
    Authors = []
    Pub_year = []
    Pub_url = []
    EEC_Author_profiles = []
    
    for page in range(0,3):
        if page == 0:
            page_no = 1
        else:
            page_url = seed+'?page='+str(page)
            page_no = page + 1
            Q.append(page_url)
                
    while(Q!=[] and count < maxcount):
        count +=1
        next_url = Q[0]
        Q = Q[1:]
                
        page_soup = make_soup(next_url)
        publications = page_soup.find_all("div",{'class':'result-container'})
        for pub in publications:
            
            #find year
            date = pub.find('span', {'class':'date'}).text
            year = date[-4:]
            Pub_year.append(year)
            print('------------------------------------------------------------------------------------------')
            print("Publication year is:",year)
            
            #find url
            publication_url = pub.find('a',{'class':'link'}).get('href')
            Pub_url.append(publication_url)
            print("Publication url is:", publication_url)
            pub_soup = make_soup(publication_url)
            
            #find title
            Title = pub_soup.find('div', {'class':'rendering'}).text 
            Pub_title.append(Title)
            print("Title is:",Title)
            
            #find authors
            pub_authors = pub_soup.find('div', {'class':'rendering_researchoutput'}).text 
            Authors.append(pub_authors)
            print("Authors are:", pub_authors)
            
            #find profile links
            EEC_profiles = pub_soup.find_all('a', {'class':'link person'})
            person_profiles =[]
            for profile in EEC_profiles: #some papers have more than 1 EEC author
                person = profile.get('href')
                person_profiles.append(person)      
                print("EEC Author Profile link is:", person)
            EEC_Author_profiles.append(person_profiles)
            print('------------------------------------------------------------------------------------------')
    
    return(Pub_title, Authors, Pub_year, Pub_url, EEC_Author_profiles)
    
    

Pub_title, Authors, Pub_year, Pub_url, EEC_Author_profiles = crawler('https://pureportal.coventry.ac.uk/en/organisations/faculty-of-engineering-environment-computing/publications/',10)
```

```python
import pandas as pd
#Generate document ID list
DocID = list(range(1,len(Pub_title)+1))

#Dictionary of results




Results = {'DocID'           : DocID,
           'Title'           : Pub_title,
           'Authors'         : Authors,
           'Year'            : Pub_year,
           'Paper_URL'       : Pub_url,
           'EEC_Profile_URLs': EEC_Author_profiles}

df_results = pd.DataFrame(Results)

print(df_results)
```

```python
#Pre-processing task 1 - remove unicode, mentions, capitals, punctiation and doubles spaces. 
import string

Title_copy = Pub_title.copy()

titles_clean = []
for d in Title_copy:
    # Remove Unicode
    title_tmp = re.sub(r'[^\x00-\x7F]+', ' ', d)
    # Remove Mentions
    title_tmp = re.sub(r'@\w+', '', title_tmp)
    # Lowercase the document
    title_tmp = title_tmp.lower()
    # Remove punctuations
    title_tmp = re.sub(r'[%s]' % re.escape(string.punctuation), ' ', title_tmp)
    # Remove the doubled space
    title_tmp = re.sub(r'\s{2,}', ' ', title_tmp)
    titles_clean.append(title_tmp)

print(titles_clean)
```

```python
#Pre-processing task 2 - tokenizing, removing stop words and stemming

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

sw = stopwords.words('english')
ps = PorterStemmer()


filtered_titles = []
for text in titles_clean:
    tokens = word_tokenize(text)
    tmp = ""
    for w in tokens:
        if w not in sw:
            tmp += ps.stem(w) + " "
    filtered_titles.append(tmp)
    
print(filtered_titles)
```

```python
index_dic = {}

#Split filtered titles into keywords
all_keywords = []
for i in range(len(filtered_titles)):
    for each_title in filtered_titles:
        words = each_title.split( )
        all_keywords.append(words)
        
        #Make each word a key in a dictionary
        for word in words:
            if word not in index_dic.keys():
                index_dic[word] = []
                
#find the publications that have each key word
for i in range(len(df_results)) :
    pub_name_keywords = df_results.loc[i, "filtered"]
    pub_ID = df_results.loc[i, "DocID"]
    x = pub_name_keywords.split()
    
    for w in x:
        if w in index_dic:
            index_dic[w].append(pub_ID)
            
print(index_dic)
print(len(index_dic))
```
